```python
import pandas as pd
from pyecharts import Bar, Boxplot

# 读取csv数据源文件（源数据条数：32699）
File_path = '/home/ls/tmp/mongodbdump/data/qcwy.csv'
df = pd.read_csv(File_path, delimiter=",", encoding="utf-8")
df = df.drop_duplicates()    # 去重（剩余： 30968）
exp_salary = pd.DataFrame(df.salary.groupby(df['experience']).mean().round(3))    # 以工作经验为索引聚合平均薪水，后计算平均薪水列的平均值保留3位小数（无工作经验项数据与预料有偏差故取3位小数），再生成DataFrame对象

# 初始化Bar()对象
bar = Bar('工作经验——平均月薪',
         title_pos='6%')
bar.add(
    "平均月薪： ",
    exp_salary.index,
    exp_salary.salary,
    label_pos = 'inside',
    # x轴设置
    xaxis_name = '工作经验/年',
    xaxis_name_pos = 'end',
    xaxis_name_gap = 12,
    xaxis_type = 'category',
    # y轴设置
    yaxis_name = 'K/月',
    yaxis_name_pos = 'end',
    yaxis_name_gap = 12,
)
bar.render("工作经验——平均月薪 柱状图.html")

# 画出箱形图查看数据
boxplot = Boxplot("工作经验——平均月薪", title_pos = '6%')
exp_salary = df[['experience', 'salary']]
x_axis = list(set(exp_salary.experience))  # 生成不重复的工作经验索引列表
x_axis.sort()    # 对x轴类目进行排序
y_axis = []     # 声明y轴空列表
for each in x_axis:
    filtrate = exp_salary[exp_salary['experience'] == each]
    y_axis.append(list(filtrate.salary.values))
_yaxis = boxplot.prepare_data(y_axis)
# 为x轴数字元素添加’年‘字
tmp = []
for i in x_axis:
    if '无工作经验' in i:
        tmp.append(i)
    else:
        tmp.append(i + '年')
boxplot.add("最小值、下四分位数、中位数、上四分位数、最大值：",
            tmp,
            _yaxis)
boxplot.render("工作经验——平均月薪 箱形图.html")
```
