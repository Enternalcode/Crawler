```python
# 

import pandas as pd
import numpy as np
from pyecharts import Bar3D

# 读取csv格式数据源文件(源数据条数:32699)
File_path = "/home/ls/tmp/mongodbdump/data/qcwy.csv"
df = pd.read_csv(File_path, delimiter=",", encoding="utf-8")

# 移除重复值(移除重复值后数据条数：30968，重复数据条数：1731)
df = df.drop_duplicates()
# df.duplicated().value_counts() # 显示去重后的数据条数
# 选中IT热门城市，删除其余数据(28个城市)
Hot_cities = ["上海", "北京", "深圳", "广州", "杭州", "南京", "成都", "武汉", "苏州", "西安", "长沙", "重庆", "合肥", "福州", "沈阳", "郑州", "天津", "大连", "无锡", "厦门", "珠海", "东莞", "宁波", "佛山", "济南", "青岛", "昆明", "南昌"]
df = df[df.city.isin(Hot_cities)] # 筛选结果剩余数据条数： 29662
# df.info() # 查看剩余数据条数
# data = df.city.groupby(df['experience']).value_counts()
# df_data = pd.DataFrame(data)    # 转化成DataFrame类型
# df_data.rename(columns={'city': 'num'}, inplace=True)    # 修改列名称为num
# df_data
data = pd.DataFrame(df.salary.groupby([df['city'], df['experience']]).mean().round(1))    # 以城市，工作经验为索引聚合平均薪水列，然后计算平均值，四舍五入取1位小数，再用DataFrame()生成DataFrame对象
# 重新设置索引
data = data.reset_index()
```

```python
# 将类目转换成数值，构造三维数组
x_axis = list(set(data.iloc[:,0]))    #　set()返回不重复元素集，list()将其变成列表
y_axis = list(set(data.iloc[:,1]))
data_xyz = []
for i in range (len(data)):
    # 以位置索引实现类目to数值的转换
    x = x_axis.index(data.iloc[i,0])    # list.index()返回第一个匹配项的位置索引，iloc[]固定列选取行
    y = y_axis.index(data.iloc[i,1])
    z = data.iloc[i,2]
    data_xyz.append([x, y, z])

bar3d = Bar3D("城市-工作经验-平均薪水", width=1200, height=1200)
range_color = ['#313695', '#4575b4', '#74add1', '#abd9e9', '#e0f3f8', '#ffffbf',
               '#fee090', '#fdae61', '#f46d43', '#d73027', '#a50026']
bar3d.add(
    "3D柱状图",
    x_axis,
    y_axis,
    [[d[0], d[1], d[2]] for d in data_xyz],
    is_visualmap = True,
    visual_range = [0,100],
    visual_range_color = range_color,
    grid3d_width = 200,     # x轴长度
    grid3d_depth = 80,      # y轴长度
    # grid3d_height = 100,    # z轴长度
    grid3d_shading = "lambert",
    # 强制全显示x轴类目
    xaxis3d_interval = 0,
    # 转动开启
    is_grid3d_rotate = True,
    # 加大力度
    grid3d_rotate_speed = 40,
)
bar3d.render("城市——工作经验——平均薪水3D柱状图 加大力度版本.html")
```
