```python
# -*- coding:utf-8 -*-
import pandas as pd
import numpy as np
from pyecharts import Geo

File_path = '/home/ls/tmp/mongodbdump/data/qcwy.csv'
# 读取csv格式数据源文件(源数据条数:32699)
def read_duplicates_csv(file_path):
    df = pd.read_csv(file_path, delimiter=",", encoding="utf-8")
    df = df.drop_duplicates()
    # print('去重后数据条数： %s' %df.duplicated().value_counts())
    # 去除形似广东省一类的异常格式数据
    # df.city.str.contains('省').value_counts() # 调试中用于查看结果
    df = df[~df.city.str.contains('省')]    # “～”取反选中不包含“省”字条目，重新赋值实现删除
    # print(df.city.value_counts())    # 剩余城市名称个数：174
    df = df.replace('开发区', '', regex = True)    # 处理“燕郊开发区”格式的数据
    return df


df = read_duplicates_csv(File_path) # 去重后剩余30968条数据
city_numjob = df.city.value_counts() # 统计每个城市的职位数量
_index = [i for i in city_numjob.index]
value = [i for i in city_numjob.values]
list_city_numjob = list(zip(_index, value))   # zip()返回一个对象(python3返回对象，python2返回元组列表)

geo = Geo(
    "全国职位分布散点图",
    title_color = '#fff',
    # 标题居中
    title_pos = 'center',
    width = 1200,
    height = 600,
    background_color = '#404a59',
)
# attr, v = geo.cast(list_city_numjob)    # 可用于处理元组列表;类似enumerate()函数，提取列表中每个元组的索引和数值;且内部为按位置提取数值

# 添加地理位置
geo.add_coordinate('怒江', 98.85, 28.85)    # 云南怒江
geo.add_coordinate('伊犁', 81.32, 43.92)    # 新疆维吾尔伊犁
geo.add_coordinate('池州', 117.48, 30.67)   # 安徽池州
geo.add_coordinate('来宾', 117.48, 30.67)   # 广西来宾
geo.add_coordinate('达州', 107.5, 31.22)    # 四川达州
geo.add_coordinate('迪庆', 99.7, 27.83)     # 云南迪庆
geo.add_coordinate('德宏', 98.58, 24.43)    # 云南德宏
geo.add_coordinate('临沧', 100.08, 23.88)   # 云南临沧
geo.add_coordinate('燕郊', 39.57, 116.51)   # 河北燕郊
geo.add_coordinate('海西', 37.37, 97.37)   # 青海海西

geo.add(
    "",
    attr,
    value,
    visual_range=[100, 7200],
    visual_text_color = '#fff',
    symbol_size = 15,
    is_visualmap = True,
    is_piecewise = True,
    visual_splir_number = 6,
)
geo.render('全国职位分布散点图_分段.html')
```
