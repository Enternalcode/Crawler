```python
# -*- coding:utf-8 -*-
import pandas as pd
from pyecharts import Bar

# 读取csv格式数据源文件(源数据条数：32699)
File_path = '/home/ls/tmp/mongodbdump/data/qcwy.csv'
df = pd.read_csv(File_path, delimiter=",", encoding="utf-8")

# 移除重复值(移除重复值后数据条数：30968，重复数据条数：1731)
df = df.drop_duplicates()
# df.duplicated().value_counts() # 显示去重后的数据条数

# 计算全国平均月薪
# salary_groupby_city = df['salary'].groupby(df['city'])    # 以城市为索引聚合平均薪水信息
# salary_groupby_city.mean()    # 查看各个城市的平均月薪
# salary_groupby_city.count()    # 查看全国平均月薪计算过程包含的样本数
avg_salary_nation = pd.Series(data = {'平均值':df.salary.mean()})   # 实例化Series()对象，传递salary列非空数据(mean()会自动跳过空值)的平均值(全国平均月薪)，用于插入数据

# 选中IT热门城市，删除其余数据(28个城市)
Hot_cities = ["上海", "北京", "深圳", "广州", "杭州", "南京", "成都", "武汉", "苏州", "西安", "长沙", "重庆", "合肥", "福州", "沈阳", "郑州", "天津", "大连", "无锡", "厦门", "珠海", "东莞", "宁波", "佛山", "济南", "青岛", "昆明", "南昌"]
df = df[df.city.isin(Hot_cities)] # 筛选结果剩余数据条数： 29662
# df.info() 查看剩余数据条数

# 添加全国平均月薪到数据中
salary_groupby_city = df['salary'].groupby(df['city']) # 以城市为索引聚合薪水信息
after_add_df = salary_groupby_city.mean().append(avg_salary_nation) # 计算各城市平均月薪后，添加全国平均月薪
sorted_df = after_add_df.sort_values(ascending=False).round(1) # sort_values(ascending=Fasle)按降序进行排列，round(1)保留1位小数，即精确到百位

_index = [i for i in sorted_df.index]
value = [i for i in sorted_df.values]
# 初始化Bar()对象设置标题
bar = Bar('城市——平均月薪',
         title_pos='6%')
bar.add('平均月薪', 
        _index, 
        value, 
        # 添加x轴 dataZoom控制条
        is_datazoom_show=True, 
        datazoom_type='both',
        # 添加y轴 dataZoom控制条
        is_datazoom_extra_show=True,
        datazoom_extra_type='both',
        # 设置x轴刻度间隔
        xaxis_interval=0,
        # 可修改柱状的颜色
        # label_color=['#eee', '#000']，
        # 内嵌显示y
        label_pos='inside',
        # x轴配置
        xaxis_name = '城市',
        xaxis_name_pos = 'end',
        xaxis_name_gap = 12,
        xaxis_type = 'category',
        # y轴配置
        yaxis_name =' K/月',
        yaxis_name_pos = 'end',
        yaxis_name_gap = 12,
       )
bar.render('城市_月薪柱状图.html')
```
